```python
import os
import support
import numpy as np

def rottentomato_dataset():
    root = 'Output/tomato_scores/'
    
    df_list = []
    for file in os.listdir(root):
        tmp = pd.read_csv(root + file)
        df_list.append(tmp)
    
    concat_df = pd.concat(df_list, axis=0).drop(columns=['Unnamed: 0']).drop_duplicates().reset_index(drop=True)
    concat_df = support.standardize_nan(concat_df)
    concat_df = concat_df.replace(0, np.nan)
    
    return concat_df
    
rotten_df = rottentomato_dataset()
rotten_df.to_csv('Output/rottentomatoes.csv')

    
```
