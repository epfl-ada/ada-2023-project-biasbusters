```python
imdb_title_basics = pd.read_csv('Data/imdb/title.basics.tsv', sep='\t')
imdb_title_ratings = pd.read_csv('Data/imdb/title.ratings.tsv', sep='\t')
imdb_df = pd.merge(left=imdb_title_basics, right=imdb_title_ratings, how='inner', on='tconst')

imdb_df = imdb_df[imdb_df['titleType'] == 'movie'].reset_index(drop=True)

# Standardize nan-values
values_to_replace = ['{}', '[]', '', 'NA', 'N/A', '-', 'nan', '\\N']
replace_map = {value: np.nan for value in values_to_replace}
imdb_df = imdb_df.replace(replace_map)

imdb_df = imdb_df[~imdb_df['genres'].isna()].reset_index(drop=True)
imdb_df['genres'] = imdb_df['genres'].apply(lambda x: str(x).split(',') if not pd.isna(x) else np.nan)


imdb_df = imdb_df[['tconst', 'titleType', 'primaryTitle', 'isAdult', 'startYear', 'runtimeMinutes', 'genres', 'averageRating', 'numVotes']]
imdb_df.columns = ['imdb_id', 'title_type', 'movie_name', 'is_adult', 'release_year', 'runtime_minutes', 'genres', 'avg_rating', 'num_votes']

imdb_df
```

```python
def compute_similarity_normalized(genre_list1, genre_list2):
    if not isinstance(genre_list1, list) or not isinstance(genre_list2, list):
        return np.nan
    
    def transform_list(input_):
        trans1 = [word for x in input_ for word in x.split(' ')]
        trans2 = [x.lower() for x in trans1]
        
        return trans2
    
    set1 = set(transform_list(genre_list1))
    set2 = set(transform_list(genre_list2))
    
    # Calculate Jaccard similarity
    intersection = set1.intersection(set2)
    union = set1.union(set2)
    jaccard_similarity = len(intersection) / len(union)

    # Normalize by dividing by the average size of the two sets
    average_size = (len(set1) + len(set2)) / 2
    normalized_similarity = jaccard_similarity / average_size
    
    return normalized_similarity
```

```python
merge_df['jaccard_similarity'] = merge_df.apply(lambda row: compute_similarity_normalized(row['genre_list'], row['genres_y']), axis=1)

merge_df
```

```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import DBSCAN
from collections import Counter

# Sample genre lists (replace with your actual data)
list_1 = ['Action', 'Comedy', 'Drama', 'Sci-Fi', 'Adventure', 'Horror']
list_2 = ['Action', 'Comedy', 'Drama']

# Create TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Combine both genre lists into a single list
combined_genres = list_1 + list_2

# TF-IDF Vectorization
tfidf_matrix = tfidf_vectorizer.fit_transform(combined_genres)

# Number of genres in list_1
num_genres_list1 = len(list_1)

# DBSCAN Clustering
dbscan = DBSCAN(eps=0.3, min_samples=2, metric='cosine')
cluster_labels = dbscan.fit_predict(tfidf_matrix)

# Extract cluster assignments for list_1
cluster_labels_list1 = cluster_labels[:num_genres_list1]

# Map list_1 genres to list_2 genres
mapping = {}
for i, genre in enumerate(list_1):
    if genre not in mapping:
        # Find all genres in list_1 with the same cluster label
        cluster_label = cluster_labels_list1[i]
        cluster_indices = [j for j, label in enumerate(cluster_labels_list1) if label == cluster_label]
        
        # Count the occurrences of each genre in list_2 within the cluster
        cluster_counts = Counter([list_2[j - num_genres_list1] for j in cluster_indices if j >= num_genres_list1])
        
        # Check if cluster_counts is empty
        if cluster_counts:
            # Find the most common genre in list_2 within the cluster
            most_common_genre = max(cluster_counts, key=cluster_counts.get)
        else:
            # If cluster_counts is empty, provide a default mapping
            most_common_genre = "Unknown"
        
        # Map the genre to the most common genre in list_2 within the cluster
        mapping[genre] = most_common_genre

# Print the mapping
for genre_list1, genre_list2 in mapping.items():
    print(f"'{genre_list1}' in list_1 maps to '{genre_list2}' in list_2")
```
