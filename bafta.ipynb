```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Define the base URL
base_url = "https://awards.bafta.org/award/{}/film"

# Define the specified awards
awards = [
    "Film | Adapted Screenplay",
    "Film | Alexander Korda Award for the outstanding British Film of the Year",
    "Film | Film",
    "Film | Film Not in the English Language",
    "Film | Original Screenplay",
    "Film | Short Animation",
    "Film | Short Film"
]

# Create an empty list to store the data
data = []

# Loop through the specified years
for year in range(1994, 2003):
    url = base_url.format(year)
    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.content, "html.parser")

        # Find all list items with class 'search-result-wrapper'
        search_results = soup.find_all("li", class_="search-result-wrapper")

        for result in search_results:
            # Extract award category and remove the year
            award_elem = result.find("div", class_="search-result-title").find("h2").find("a")
            award = award_elem.text.replace(f" in {year}", "")

            # Check if the extracted award is in the specified awards list
            if award in awards:
                # Find all headlines within each search result
                all_headlines = result.find_all("div", class_="search-result-headline")

                for headline in all_headlines:
                    # Extract movie name
                    movie_name = headline.find("p").get_text(strip=True)

                    # Check if the movie name is empty; if so, skip this entry
                    if not movie_name:
                        continue

                    # Check if the headline includes 'search-result-winner' class
                    win_status = 'True' if 'search-result-winner' in headline['class'] else 'False'

                    # Extract people and split into a list
                    people = headline.find_next_sibling("div", class_="search-result-subtitle").find("p").get_text(strip=True)
                    people = people.split(', ')

                    # Check if the award is 'Alexander Korda Award for the outstanding British Film of the Year'
                    # If so, set country as 'Great Britain'
                    if award == "Film | Alexander Korda Award for the outstanding British Film of the Year":
                        country = 'Great Britain'
                    else:
                        country = None

                    # Append data to the list
                    row_data = {
                        'year_film': year - 1,
                        'year_award': year,
                        'ceremony': 11 + year - 1994,  # Calculating ceremony based on the year
                        'category': award,
                        'win': win_status,
                        'movie_name': movie_name,
                        'people': people,
                        'country': country,
                        'award': 'BAFTA'
                    }

                    # Append data to the list if the award matches and movie name is not empty
                    data.append(row_data)

# Create a DataFrame from the collected data
df = pd.DataFrame(data)

# Display the DataFrame
df
```
